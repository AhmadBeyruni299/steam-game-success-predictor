**Handling Outliers**

There appears to be a number of outliers in the dataset. This is expected since the major titles/games in Steam is likely to skew the data in several aspects

For example: Games like CS2/Dota2 level games skewing the review since they're very popular. These outliers are legitimate games (CS2, Dota 2 etc). They're not data errors. Removing them would introduce bias. 

Solution: Log transform skewed columns

```python
df['price_log'] = np.log1p(df['price'])
df['positive_log'] = np.log1p(df['positive'])
df['playtime_log'] = np.log1p(df['average_playtime_forever'])
df['peak_ccu_log'] = np.log1p(df['peak_ccu'])

# Visualise the difference
fig, axes = plt.subplots(2, 4, figsize=(20, 8))

df['price'].hist(ax=axes[0,0], bins=50)
axes[0,0].set_title('Price (original)')

df['price_log'].hist(ax=axes[0,1], bins=50)
axes[0,1].set_title('Price (log)')

df['positive'].hist(ax=axes[0,2], bins=50)
axes[0,2].set_title('Positive Reviews (original)')

df['positive_log'].hist(ax=axes[0,3], bins=50)
axes[0,3].set_title('Positive Reviews (log)')

df['playtime_log'].hist(ax=axes[1,0], bins=50)
axes[1,0].set_title('Playtime (log)')

df['peak_ccu_log'].hist(ax=axes[1,1], bins=50)
axes[1,1].set_title('Peak CCU (log)')

plt.tight_layout()
plt.show()
```

**Genres**

```python
# Extract individual genres and calculate success rate per genre
from collections import Counter

# Split genre strings and explode into individual rows
df['genres_list'] = df['genres'].str.strip("[]").str.replace("'", "").str.split(", ")
genres_exploded = df.explode('genres_list')

genre_success = genres_exploded.groupby('genres_list')['success'].agg(['mean', 'count'])
genre_success = genre_success[genre_success['count'] > 100].sort_values('mean', ascending=False)

plt.figure(figsize=(14, 7))
genre_success['mean'].plot(kind='bar', color='steelblue')
plt.title('Success Rate by Genre')
plt.xlabel('Genre')
plt.ylabel('Success Rate')
plt.axhline(y=df['success'].mean(), color='red', linestyle='--', label='Overall average')
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Genre shows a meaningful relationship with success. Casual, Indie and Adventure games perform at or above the average success rate while Gore, Violent and Massively Multiplayer games significantly underperform. Early Access games also struggle, likely due to player expectations of unfinished products.

```python
from itertools import combinations

# Get games with more than one genre
df['genres_list'] = df['genres'].str.strip("[]").str.replace("'", "").str.split(", ")
multi_genre = df[df['genres_list'].apply(lambda x: len(x) > 1)]

print(f"Games with multiple genres: {len(multi_genre)}")
print(f"Single genre games: {len(df) - len(multi_genre)}")

# Get all genre pairs and their success rates
pair_success = {}

for _, row in multi_genre.iterrows():
    for pair in combinations(sorted(row['genres_list']), 2):
        if pair not in pair_success:
            pair_success[pair] = []
        pair_success[pair].append(row['success'])

# Convert to dataframe
pair_df = pd.DataFrame([
    {'genre_pair': f"{a} + {b}", 
     'success_rate': sum(v)/len(v), 
     'count': len(v)}
    for (a, b), v in pair_success.items()
])

# Filter for pairs with enough games to be meaningful
pair_df = pair_df[pair_df['count'] > 50].sort_values('success_rate', ascending=False)

print(f"\nTotal meaningful genre pairs found: {len(pair_df)}")
print(pair_df.head(10))
```

```python
# Plot top and bottom genre pairs
fig, axes = plt.subplots(2, 1, figsize=(14, 12))

# Top 15 performing pairs
top_pairs = pair_df.head(15)
axes[0].barh(top_pairs['genre_pair'], top_pairs['success_rate'], color='steelblue')
axes[0].axvline(x=df['success'].mean(), color='red', linestyle='--', label='Overall average')
axes[0].set_title('Top 15 Genre Pairs by Success Rate')
axes[0].set_xlabel('Success Rate')
axes[0].legend()
axes[0].invert_yaxis()

# Bottom 15 performing pairs
bottom_pairs = pair_df.tail(15)
axes[1].barh(bottom_pairs['genre_pair'], bottom_pairs['success_rate'], color='salmon')
axes[1].axvline(x=df['success'].mean(), color='red', linestyle='--', label='Overall average')
axes[1].set_title('Bottom 15 Genre Pairs by Success Rate')
axes[1].set_xlabel('Success Rate')
axes[1].legend()
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()
```

Pairwise genre analysis reveals that no combination significantly outperforms the overall average, suggesting genre alone is insufficient to predict success. However, certain combinations are strongly negative predictors like Gore, Massively Multiplayer and Violent content consistently produce the lowest success rates regardless of pairing. Indie appears as a positive signal across nearly all combinations.

```python
# Look at a few raw tag values
print(df['tags'].iloc[0])
print("\n---\n")
print(df['tags'].iloc[1])
print("\n---\n")
print(df['tags'].iloc[2])
```

There are simply too many unique tags to be listed down. Tags were community applied so there are likely hundreds of unique values

```python
import ast

def extract_top_tags(tag_string, top_n=5):
    try:
        tag_dict = ast.literal_eval(tag_string)
        # Sort by vote count and take top N tags
        sorted_tags = sorted(tag_dict.items(), key=lambda x: x[1], reverse=True)
        return [tag for tag, count in sorted_tags[:top_n]]
    except:
        return []

# Extract top 5 tags per game
df['tags_list'] = df['tags'].apply(extract_top_tags)

# Check it worked
print(df['tags_list'].head(5))
```

```python
all_tags = [tag for tags in df['tags_list'] for tag in tags]
tag_counts = Counter(all_tags)
print(f"Unique tags after filtering: {len(tag_counts)}")
print(f"\nTop 20 most common tags:")
for tag, count in tag_counts.most_common(20):
    print(f"{tag}: {count}")
```

After analysing the most common tags and filtering the unique tags, I decided to only keeping tags that appear in 500+ games to keep the number of columns managable 

```python
common_tags = [tag for tag, count in tag_counts.items() if count >= 500]
print(f"Tags appearing in 500+ games: {len(common_tags)}")

# Create one-hot encoding for common tags only
for tag in common_tags:
    df[f'tag_{tag}'] = df['tags_list'].apply(lambda x: 1 if tag in x else 0)

tag_cols = [col for col in df.columns if col.startswith('tag_')]


print(f"\nDataframe shape after tag encoding: {df.shape}")
print(f"Number of tag features added: {len(tag_cols)}")
print(tag_cols)
```
