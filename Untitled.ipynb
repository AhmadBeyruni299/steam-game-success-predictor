**Loading the data:**

The dataset was obtained through Kaggle named 'Steam Games Dataset' published by Artemiy Ermilov et al. It contains a total of 88,899 entries last updated in 2024

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries loaded successfully!")
```

```python
df = pd.read_csv("games_march2025_cleaned.csv")
df.head()
```

**Initial observation:**

The dataset contained 47 columns. Some of which had a lot of missing of data. Several columns like Movies and Score rank were too sparse to be useful and were dropped.

```python
df.info()
```

```python
columns_dropped = ['header_image', 'website', 'support_url', 'support_email', 'metacritic_url', 'screenshots', 'notes']

df = df.drop(columns= columns_dropped)
df.head()
```

```python
print(f"Remaining columns: {df.shape[1]}")
```

**Inspecting missing values**

Score rank and reviews had over 89000 and 79000 missing data respectively leading it to be deemed as statistically irrelevant for this project. Therefore, these columns was removed

detailed_description , about_the_game and short_description is dropped. The scope of the project is deemed too complicated if these columns were to be taken into account using NLP. Hence, predicting success through NLP using these columns will be a future update instead

Games without reviews are dropped with the conclusion 

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['score_rank', 'reviews', 'detailed_description', 'about_the_game', 'short_description'])

print(df.shape)
```

```python
#Checking for 'Ghost games: Games without reviews'
print(f"pct_pos_total == -1: {(df['pct_pos_total'] == -1).sum()}")
print(f"num_reviews_total == -1: {(df['num_reviews_total'] == -1).sum()}")
```

```python
df = df[df['pct_pos_total'] != -1]
print(df.shape)
```

**Defining success**

Games with no owners were removed from the dataset. Without any owner, eve if there's reviews meant that the data is faulty and unreliable

Success in the context of this project is defined as 70%+ positive reviews 

```python
# Check the 0 - 0 and 0 - 20000 buckets
print(df['estimated_owners'].value_counts())

# How many games have '0 - 0' owners?
print(f"\nGames with 0-0 owners: {(df['estimated_owners'] == '0 - 0').sum()}")

# How many games have '0 - 20000' owners?
print(f"Games with 0-20000 owners: {(df['estimated_owners'] == '0 - 20000').sum()}")
```

```python
df = df[df['estimated_owners'] != '0 - 0']
print(df.shape)
```

```python
#Define the success condition for a game
df['success'] = (df['pct_pos_total'] >= 70).astype(int)

print(df['success'].value_counts())
print(df['success'].value_counts(normalize=True))
```
